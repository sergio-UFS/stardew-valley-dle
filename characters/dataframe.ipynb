```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
df = pd.read_csv('characters.csv')
df = df[['Name','Birthday Season','Birthday Day','Lives In','Address','Family','Marriage','Clinic Visit','Loved Gifts','Liked Gifts','Neutral Gifts','Disliked Gifts','Hated Gifts']]
df = df.fillna('')

df['Family'] = df['Family'].apply(lambda x: [] if x == '' else x)
df['Clinic Visit']= df['Clinic Visit'].apply(lambda x: 'Never' if x == '' else x)

df.to_csv('characters.csv', index=False)
print(tabulate(df, headers='keys', tablefmt='psql'))
```
